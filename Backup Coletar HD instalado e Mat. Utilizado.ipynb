```python
#ACESSO AO BANCO DE DADOS
def executar_conexao_bd():
    server = 'TBDBL-GLEISSON'
    database = 'Dados_Pavimento'
    username = 'gleisson.santos'
    password = '1234'
    conexao = pyodbc.connect(
        'DRIVER={SQL Server};SERVER=' + server + ';DATABASE=' + database + ';UID=' + username + ';PWD=' + password)
    print('BD atualizado com sucesso!! :) ')
    return conexao
```

```python
#FUNÇÃO PRINCIPAL QUE ME RETORNA VALORES DOS TITLES
def preencher_ss(driver, lista, lista2):
    driver.find_element(By.XPATH, '//*[@id="form-j_idt70-j_idt71-item"]').clear()
    driver.find_element(By.XPATH, '//*[@id="form-j_idt70-j_idt79-item"]').clear()
    
    driver.find_element(By.XPATH, '//*[@id="form-j_idt70-j_idt71-item"]').send_keys(lista)
    time.sleep(0.3)
    driver.find_element(By.XPATH, '//*[@id="form-j_idt70-j_idt79-item"]').send_keys(lista2)

    # clinical em buscar
    driver.find_element(By.XPATH, '//*[@id="form-j_idt70-panel"]/input').click() 
    try:
        HD = driver.find_element(By.CSS_SELECTOR, "label[title='Número do Hidrômetro Instalado'] ").text                                                                  
    except:
        HD = driver.find_element(By.CSS_SELECTOR, "label[title='Hidrômetro Instalado'] ").text
    
    
    Material = driver.find_element(By.CSS_SELECTOR, "label[title='Material Utilizado']").text
    
    return HD, Material
```

```python
#IMPORTANDO PLANILHA DO BANCO DE DADOS PARA CONSULTA
df = pd.read_sql('SELECT TOP 50 *  FROM Material', executar_conexao_bd())
df = pd.DataFrame(df, columns=['SS', 'OS', 'Matutilizado', 'Hdutilizado', 'Usoumureta'])
df = df.astype(str)
df
```

```python
#FAZER CONSULTA AO ITEM DO SISTEMA UTILIZANDO A FUNÇÃO / ATUALIZANDO DUAS COLUNAS
HD = []
Mat= []
for i, row in df.iterrows():
    a = preencher_ss(driver, row['SS'], row['OS']) #utilizando a função preencher_ss
    NumHD = a[0]
    Mater = a[1]
    Mat.append(Mater)
    HD.append(NumHD)
    
df['Matutilizado'] = Mat
df['Hdutilizado'] = HD
df
```

```python
# UTILIZANR ITEM DA COLUNA Matutilizado E VERIFICAR SE POSSUI UM TEXTO ESPECIFICO
m_encontrado = ['capsulame']
material = []
for i, row in df.iterrows():
    if any(x in row['Matutilizado'] for x in m_encontrado):#retornar True
        material.append('Sim, 1Unid')
    else:
        material.append('Não utilizou')
df['Usoumureta'] = material
df
```

```python
df = df.astype(str)
df = pd.DataFrame(df, columns=['SS', 'OS', 'Matutilizado', 'Hdutilizado', 'Usoumureta'])

cursor = executar_conexao_bd().cursor()
for index, row in df.iterrows():
    Mat = row.Matutilizado
    HD  = row.Hdutilizado
    INS = row.Usoumureta
    SS  = row.SS
    comando_sql = (f"UPDATE Material SET Matutilizado='{Mat}', Hdutilizado='{HD}', Usoumureta='{INS}' WHERE SS ={SS}")
    cursor.execute(comando_sql)
    
cursor.commit()
cursor.close()
```
